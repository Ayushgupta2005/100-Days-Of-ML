# Distance Metrics Visualization
**Goal:**
1. Understand how KNN measures similarity
2. Visualize Euclidean Distance
3. Visualize Manhattan Distance
4. Understand why nearest neighbors matter

```python
# Import required libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating Two sample points

A = np.array([2, 3])

B = np.array([7, 8])

print("Point A:", A)
print("Point B:", B)
```

```python
# Plot points

plt.figure(figsize=(6,6))

plt.scatter(A[0],A[1],s=150,label="A"
)

plt.scatter(B[0],B[1],s=150,label="B"
)

plt.plot(
    [A[0], B[0]],
    [A[1], B[1]]
)

plt.xlim(0,10)
plt.ylim(0,10)

plt.grid()

plt.legend()

plt.title("Two Points in 2D Space")

plt.show()
```

```python
# Euclidean Distance Formula

euclidean_distance = np.sqrt(
    np.sum((A - B)**2)
)

print(
    "Euclidean Distance:",
    round(euclidean_distance, 2)
)
```

## Euclidean Distance

Straight line distance between two points.

Formula:

distance = √((x₂-x₁)² + (y₂-y₁)²)

```python
# Manhattan Distance Formula

manhattan_distance = np.sum(
    np.abs(A - B)
)

print(
    "Manhattan Distance:",
    manhattan_distance
)
```

## Manhattan Distance

Distance travelled only horizontally and vertically.

Formula:

distance = |x₂-x₁| + |y₂-y₁|

```python
# Comparing Euclidean and Manhattan Distances

plt.figure(figsize=(6,6))

plt.scatter(A[0], A[1], s=150)
plt.scatter(B[0], B[1], s=150)

# Euclidean Path
plt.plot([A[0], B[0]],[A[1], B[1]],label="Euclidean")

# Manhattan Path
plt.plot([A[0], B[0]],[A[1], A[1]],'--')

plt.plot([B[0], B[0]],[A[1], B[1]],'--',label="Manhattan")

plt.legend()

plt.grid()

plt.title("Euclidean vs Manhattan")

plt.show()
```

**Why Distance Matter**

```python
# Sample training points

X = np.array([
    [1,1],
    [2,2],
    [3,3],
    [8,8],
    [9,9],
    [10,10]
])

y = np.array([
    0,
    0,
    0,
    1,
    1,
    1
])

new_point = np.array([4,4])
```

```python
# Visualize classes
plt.figure(figsize=(6,6))

plt.scatter(X[y==0][:,0],X[y==0][:,1],s=100,label="Class 0")

plt.scatter(X[y==1][:,0],X[y==1][:,1],s=100,label="Class 1")

plt.scatter(new_point[0],new_point[1],s=200,marker='*',label="Unknown Point")

plt.legend()

plt.grid()
plt.title("KNN Intuition")
plt.show()
```

## KNN Intuition

For a new point:

1. Calculate distance to all points.
2. Find nearest neighbors.
3. Let neighbors vote.
4. Assign the majority class.

No training is performed.

KNN learns directly from stored examples.

```python
# Distance from new point to every point

distances = []
for point in X:
    distance = np.sqrt(
        np.sum((point - new_point)**2)
    )
    distances.append(distance)

print(distances)
```

```python
# Closest point

nearest_index = np.argmin(distances)

print("Nearest Point:",X[nearest_index])

print("Class:",y[nearest_index])
```

# Conclusions

- **KNN uses distance to measure similarity.**
- **Euclidean distance is the most common metric.**
- **KNN does not learn weights.**
- **Predictions are based on nearby points.**
- **Similar points are expected to belong to the same class.**
