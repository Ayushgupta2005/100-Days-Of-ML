```python
# Day 03 - Vectors & Matrices

import numpy as np
```

```python
# Creating a vector
vector = np.array([1, 2, 3])

print("Vector:", vector)
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# Addition
print("Addition:", v1 + v2)

# Subtraction
print("Subtraction:", v1 - v2)

# Scalar multiplication
print("Scalar Multiplication:", v1 * 2)
```

```python
# Dot Product
dot = np.dot(v1, v2)

print("Dot Product:", dot)
```

```python
# Creating a matrix
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

print("Matrix:\n", matrix)
```

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])

# Addition
print("Addition:\n", m1 + m2)

# Multiplication (element-wise)
print("Element-wise:\n", m1 * m2)
```

```python
# Matrix multiplication
result = np.dot(m1, m2)

print("Matrix Multiplication:\n", result)
```

```python
# Transpose
transpose = m1.T

print("Transpose:\n", transpose)
```
