## Singular Value Decomposition (SVD)

A matrix A can be decomposed as:

A = U Σ Vᵀ

Where:
- U → left singular vectors
- Σ → diagonal matrix (singular values)
- Vᵀ → right singular vectors

```python
import numpy as np

A = np.array([[3, 1],
              [1, 3]])

print(A)
```

```python
# Step 1: Compute AᵀA
A_T = A.T
ATA = np.dot(A_T, A)

print("A^T A:\n", ATA)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(ATA)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# Singular Values
singular_values = np.sqrt(eigenvalues)

print("Singular Values:", singular_values)
```

```python
# U = (A * V) / singular values
U = np.dot(A, eigenvectors)

print("U (approx):\n", U)
```

## Understanding

- SVD breaks matrix into simpler parts
- Helps reduce dimensions
- Used in PCA, recommendation systems
