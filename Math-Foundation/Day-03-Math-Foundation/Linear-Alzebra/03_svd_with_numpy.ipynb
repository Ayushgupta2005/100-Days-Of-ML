```python
# Day 03 - SVD using NumPy

import numpy as np
```

```python
A = np.array([[3, 1],
              [1, 3]])

print("Matrix A:\n", A)
```

```python
# Applying SVD
U, S, VT = np.linalg.svd(A)

print("U:\n", U)
print("Singular Values:", S)
print("V^T:\n", VT)
```

## Output of SVD

U → left singular vectors  
S → singular values  
Vᵀ → right singular vectors  

A = U Σ Vᵀ

```python
# Convert S to diagonal matrix
S_matrix = np.diag(S)

# Reconstruct A
A_reconstructed = np.dot(U, np.dot(S_matrix, VT))

print("Reconstructed A:\n", A_reconstructed)
```

```python
print("Original:\n", A)
print("\nReconstructed:\n", A_reconstructed)
```
