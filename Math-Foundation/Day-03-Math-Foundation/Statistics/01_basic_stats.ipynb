```python
# Day 03 - Basic Statistics
# Author: Ayush Gupta

import numpy as np
import pandas as pd
```

```python
# Sample Data
data = [10, 20, 30, 40, 50]

arr = np.array(data)
print(arr)
```

```python
# Mean
mean = np.mean(arr)

print("Mean:", mean)
```

```python
# Median
median = np.median(arr)
print("Median:", median)
```

```python
# Standard Deviation
std = np.std(arr)

print("Standard Deviation:", std)
```
