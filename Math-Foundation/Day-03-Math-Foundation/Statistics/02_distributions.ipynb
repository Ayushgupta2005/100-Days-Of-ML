```python
# Day 03 - Distributions
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating normal distribution data
data = np.random.normal(loc=0, scale=1, size=1000)

print(data[:10])
```

```python
# Histogram
plt.hist(data, bins=30)

plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

## Normal Distribution

- Symmetrical around mean
- Mean = Median = Mode
- Bell-shaped curve

```python
# Right skewed data
data_right = np.random.exponential(scale=1, size=1000)

plt.hist(data_right, bins=30)

plt.title("Right Skewed Distribution")

plt.show()
```


Right Skew → tail on right  
Mean > Median  

```python
# Left skew (simple example)
data_left = -np.random.exponential(scale=1, size=1000)

plt.hist(data_left, bins=30)

plt.title("Left Skewed Distribution")

plt.show()
```

Left Skew → tail on left  
Mean < Median  

```python
print("Mean:", np.mean(data_right))
print("Median:", np.median(data_right))
```
