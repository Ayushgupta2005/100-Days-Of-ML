```python
# Day 03 - Correlation & Covariance

import numpy as np
import pandas as pd
```

```python
# Sample Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])   # perfectly related x= 2y

df = pd.DataFrame({"X": x, "Y": y})
print(df)
```

```python
# Covariance
cov = np.cov(x, y)

print("Covariance Matrix:\n", cov)
```

	•	Positive → variables increase together
	•	Negative → opposite movement

```python
# Correlation
corr = np.corrcoef(x, y)

print("Correlation Matrix:\n", corr)
```

Value range:
	•	+1 → strong positive
	•	0 → no relation
	•	-1 → strong negative

```python
# Correlation using pandas
print(df.corr())
```

```python
# Negative relation Example
y2 = np.array([10, 8, 6, 4, 2])

print(np.corrcoef(x, y2))
```

```python
# Random data . No Relationship
y3 = np.random.rand(5)

print(np.corrcoef(x, y3))
```

```python
# With Real Datasetx
df = pd.read_csv("data.csv")

print(df.corr(numeric_only=True))
```

## Why Correlation & Covariance?

- Helps understand relationships between features
- Useful for feature selection
- Avoids redundant features
