```python
# Day 01 - NumPy Basics
# Author: Ayush Gupta
# Description: Learning NumPy for numerical operations in Machine Learning

import numpy as np
```

```python
# Creating ARRays
# 1D Array
arr1 = np.array([1, 2, 3, 4])
print(arr1)
```

```python
# 2D Array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
```

```python
# Using built-in functions
zeros = np.zeros((2,3))
ones = np.ones((2,2))
identity = np.eye(3)

print(zeros)
print(ones)
print(identity)
```

```python
# Attributes of Arrays
arr = np.array([[1,2,3],[4,5,6]])

print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
print("Size:", arr.size)
print("Data Type:", arr.dtype)
```

```python
# Indexing and Slicing
arr = np.array([10, 20, 30, 40, 50])

print(arr[0])        # First element
print(arr[-1])       # Last element
print(arr[1:4])      # Slicing

# 2D indexing
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2[0,1])     # Row 0, Column 1
```

```python
# Array Operation
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
```

```python
print(a - b)
```

```python
print(a * b)
```

```python
print(a / b)
```

```python
# Array Broadcasting
arr = np.array([1,2,3])
print(arr + 5)   # Adds 5 to each element
```

```python
# Important Functions
arr = np.array([1,2,3,4,5])

print(np.mean(arr))
print(np.sum(arr))
print(np.max(arr))
print(np.min(arr))
print(np.std(arr))
```

```python
# Array Reshaping
arr = np.array([1,2,3,4,5,6])

reshaped = arr.reshape(2,3)
print(reshaped)
```

```python
# Random Numbers
rand = np.random.rand(3,3)        # Random between 0 and 1
rand_int = np.random.randint(1,10,(2,3))

print(rand)
print(rand_int)
```

```python
# Conditional Selection 
arr = np.array([10, 20, 30, 40])

print(arr > 20)
print(arr[arr > 20])
```

```python
arr = np.array([1,2,3])

copy_arr = arr.copy()
view_arr = arr.view()

arr[0] = 99

print("Original:", arr)
print("Copy:", copy_arr)
print("View:", view_arr)
```
