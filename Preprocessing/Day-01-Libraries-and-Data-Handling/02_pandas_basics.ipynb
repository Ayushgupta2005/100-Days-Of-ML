```python
# Day 01 - Pandas Basics
# Author: Ayush Gupta
# Description: Learning Pandas for data handling and analysis

import pandas as pd
```

```python
# Creating DataFrame
data = {
    "Name": ["A", "B", "C"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Loading Dataset from CSV
df = pd.read_csv("data.csv")

print(df.head())
```

```python
# Viewing First and Last Rows
print(df.head())
print(df.tail())
```

```python
# Dataset Information
print(df.info())
```

```python
# Statistical Summary
print(df.describe())
```

```python
# Shape and Columns
print(df.shape)
print(df.columns)
```

```python
# Selecting Single Column
print(df[" income_annum"])
```

```python
# Checking Missing Values
print(df.isnull().sum())

# Filling Missing Values
df.fillna(0, inplace=True)

# Dropping Missing Values
df.dropna(inplace=True)
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
# Sorting by Marks (Descending)
df_sorted = df.sort_values(by="income_annum", ascending=False)

print(df_sorted)
```
