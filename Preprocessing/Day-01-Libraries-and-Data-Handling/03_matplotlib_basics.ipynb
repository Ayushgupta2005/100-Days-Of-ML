```python
# Day 01 - Pandas Basics
# Author: Ayush Gupta
# Description: Learining matplotlib for data Visualization

# Import Matplotlib
import matplotlib.pyplot as plt
```

```python
# Creating Data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Line Plot
plt.plot(x, y)

# Show Plot
plt.show()
```

```python
# Line Plot with Labels
plt.plot(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")

plt.show()
```

```python
# Scatter Plot
plt.scatter(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")

plt.show()
```

```python
# Bar Chart
categories = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(categories, values)

plt.title("Bar Chart")

plt.show()
```

```python
# Pie Chart
labels = ["A", "B", "C"]
sizes = [40, 35, 25]

plt.pie(sizes, labels=labels)

plt.title("Pie Chart")

plt.show()
```

```python
# Custom Line Plot
plt.plot(x, y, marker='o', linestyle='--')

plt.title("Customized Plot")
plt.xlabel("X")
plt.ylabel("Y")

plt.grid()

plt.show()
```

```python
# Multiple Lines
y2 = [15, 18, 22, 28]

plt.plot(x, y)
plt.plot(x, y2)

plt.title("Multiple Lines")

plt.show()
```

```python
# Histogram
data = [1,2,2,3,3,3,4,4,5]

plt.hist(data)

plt.title("Histogram")

plt.show()
```

```python
# Practice Plot
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y)
plt.scatter(x, y)

plt.title("Line + Scatter")
plt.show()
```
