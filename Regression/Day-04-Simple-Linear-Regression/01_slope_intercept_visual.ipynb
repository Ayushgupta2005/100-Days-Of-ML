```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])
```

```python
# y = mx + c (Equation of line)
m = 2   # slope
c = 0   # intercept

y_pred = m * x + c
```

```python
plt.scatter(x, y, label="Actual Data")
plt.plot(x, y_pred, color='red', label="Line")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = mx + c")

plt.legend()
plt.show()
```

```python
m = 1
c = 0 #Slope changed

y_pred = m * x + c

plt.scatter(x, y)
plt.plot(x, y_pred, color='green')

plt.title("Different Slope")
plt.show()
```

```python
# Intercept changed
m = 2
c = 2

y_pred = m * x + c

plt.scatter(x, y)
plt.plot(x, y_pred, color='orange')

plt.title("Different Intercept")
plt.show()
```

## Understanding

- m → slope (how steep the line is)
- c → intercept (where line crosses y-axis)

Changing m → rotates line  
Changing c → shifts line up/down
