# Polynomial Regression
# Dataset: Ice Cream Sales

```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Load dataset 
df = pd.read_csv("Ice_cream selling data.csv")

print(df.head())
```

```python
#  Define X and y
X = df.iloc[:, 0].values.reshape(-1, 1)   # temperature
y = df.iloc[:, 1].values                 # sales
```

```python
# Visualize Data
plt.scatter(X, y)

plt.xlabel("Temperature")
plt.ylabel("Ice Cream Sales")
plt.title("Data Visualization")

plt.show()
```

## Observation

- Data is not a straight line  
- It has a curve pattern  

So linear regression will not fit properly  

```python
# Applying Linear Regression
lin_model = LinearRegression()
lin_model.fit(X, y)

y_lin_pred = lin_model.predict(X)

plt.scatter(X, y)
plt.plot(X, y_lin_pred, color='red')

plt.title("Linear Regression (Bad Fit)")
plt.show()
```

## Why Linear Regression Fails

- Linear regression tries to fit a straight line  
- But data is curved  
- So model underfits  

This is called underfitting  

```python
# Applying polynomial feartures
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
# Training Polynomial Model
poly_model = LinearRegression()
poly_model.fit(X_poly, y)
```

```python
y_poly_pred = poly_model.predict(X_poly) #Prediction 
```

```python
plt.scatter(X, y)

# smooth curve
X_grid = np.linspace(min(X), max(X), 100).reshape(-1, 1)
X_grid_poly = poly.transform(X_grid)

plt.plot(X_grid, poly_model.predict(X_grid_poly), color='green')

plt.title("Polynomial Regression (Good Fit)")
plt.show()
```
