# Sigmoid Function Visualization
**Goal:**
 1. Understand why Logistic Regression uses Sigmoid
 2. Visualize Sigmoid Curve
 3. Understand Probability Output
 4. Understand Decision Boundary

```python
# Import required libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid Function

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Create values from -10 to +10

z = np.linspace(-10, 10, 1000)

sigmoid_values = sigmoid(z)
```

```python
# Visualize Sigmoid Function

plt.figure(figsize=(8,5))

plt.plot(
    z,
    sigmoid_values,
    linewidth=3
)

plt.title("Sigmoid Function")

plt.xlabel("z")

plt.ylabel("σ(z)")

plt.grid()

plt.show()
```

## Observation

The sigmoid function converts any input value into a number between 0 and 1.

Very negative values approach 0.

Very positive values approach 1.

```python
# Check some important sigmoid values

values = [-10, -5, 0, 5, 10]

for v in values:
    print(
        f"Sigmoid({v}) = {sigmoid(v):.4f}"
    )
```

```python
# Highlight decision boundary

plt.figure(figsize=(8,5))

plt.plot(
    z,
    sigmoid_values,
    linewidth=3
)

plt.axhline(
    y=0.5,
    linestyle='--',
    label='Probability = 0.5'
)

plt.axvline(
    x=0,
    linestyle='--',
    label='z = 0'
)

plt.legend()

plt.title(
    "Sigmoid Decision Boundary"
)

plt.show()
```

## Decision Boundary

If probability >= 0.5:
    Predict Class 1

If probability < 0.5:
    Predict Class 0

Since:

Sigmoid(0) = 0.5

the decision boundary occurs at z = 0.

```python
# Example probabilities

examples = [-3, -1, 0, 1, 3]

for e in examples:

    probability = sigmoid(e)

    print(
        f"z={e} --> Probability={probability:.4f}"
    )
```

## Why Sigmoid?

Linear Regression can predict values greater than 1 or less than 0.

Example:

-5
20
100

These cannot represent probabilities.

Sigmoid converts any value into the range:

0 to 1

making it suitable for classification problems.

```python
# Compare raw linear output with sigmoid output

z = np.linspace(-10, 10, 100)

plt.figure(figsize=(8,5))

plt.plot(
    z,
    z,
    label="Linear Output"
)

plt.plot(
    z,
    sigmoid(z),
    label="Sigmoid Output"
)

plt.legend()

plt.title(
    "Linear Output vs Sigmoid Output"
)

plt.show()
```


- Sigmoid enables Linear Regression concepts to be used for Classification.

```python
# Example points belonging to two classes

x = np.array([-4, -3, -2, -1, 1, 2, 3, 4])

y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

plt.figure(figsize=(8,5))

plt.scatter(
    x[y==0],
    y[y==0],
    s=100,
    label="Class 0"
)

plt.scatter(
    x[y==1],
    y[y==1],
    s=100,
    label="Class 1"
)

plt.axvline(
    x=0,
    linestyle="--",
    color="red",
    label="Decision Boundary"
)

plt.xlabel("Feature")

plt.ylabel("Class")

plt.title(
    "Binary Classification Example"
)

plt.legend()

plt.show()
```

## Classification Problem

We have two classes:

Class 0 → Negative Class

Class 1 → Positive Class

The goal of Logistic Regression is to find a decision boundary that separates these classes.

```python
# Convert raw scores into probabilities

x_values = np.linspace(-6, 6, 500)

probabilities = sigmoid(x_values)

plt.figure(figsize=(8,5))

plt.plot(
    x_values,
    probabilities,
    linewidth=3
)

plt.axhline(
    y=0.5,
    linestyle="--",
    color="red"
)

plt.axvline(
    x=0,
    linestyle="--",
    color="red"
)

plt.xlabel("Raw Score (z)")
plt.ylabel("Probability")

plt.title(
    "Decision Boundary Using Sigmoid"
)

plt.show()
```

## What Is Happening?

Raw model output:

z = w₁x₁ + w₂x₂ + ... + b

can be any value:

-∞ to +∞

Sigmoid converts it into:

0 to 1

which can be interpreted as a probability.

```python
# Example predictions

test_values = [-4, -2, -1, 0, 1, 2, 4]

for value in test_values:

    probability = sigmoid(value)

    prediction = 1 if probability >= 0.5 else 0

    print(
        f"z={value:>2} | Probability={probability:.4f} | Predicted Class={prediction}"
    )
```

Decision Rule:

Probability >= 0.5 → Class 1

Probability < 0.5 → Class 0
